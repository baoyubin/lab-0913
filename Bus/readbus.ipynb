```python
import geopandas as gpd
#导入TransBigData包
import transbigdata as tbd
#读取数据
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bounds = [121.4,31.15,121.55,31.35]
#bounds = [121.0, 31.0, 122.0, 32.0]
params = tbd.area_to_params(bounds, accuracy=500, method='rect')
params
```

```python
data = pd.read_csv('bus_vehicleNum/04S2C080.txt')
#01S0D016.txt (01S0D034 01S0D039 01S0D040 )(01S0D041 42 45 47 48)(105 106 110)  109
data = tbd.clean_outofbounds(data,bounds = bounds,col = ['lng','lat'])
data['LONCOL'], data['LATCOL'] = tbd.GPS_to_grid(data['lng'], data['lat'], params)
dataset = data.groupby(['LONCOL', 'LATCOL'])['vehicleNum'].count().reset_index()
grid = [dataset['LONCOL'], dataset['LATCOL']]
dataset['geometry'] = tbd.grid_to_polygon(grid, params)
datatest = gpd.GeoDataFrame(dataset)
datatest.plot(column='vehicleNum',cmap='Blues')
```
